```python
import tensorflow as tf
```

```python
#从均匀分布中输出随机值
img = tf.random_uniform([10,28,28,3])
with tf.Session() as sess1:
    print(sess1.run(img))
```

```python
#利用索引取出第二张图片
a = img[1:2,0:,0:,0:]
with tf.Session() as sess2:
    print(sess2.run(a))
```

```python
#利用切片取出第二张图片
b = tf.slice(img,[1,0,0,0],[1,28,28,3])
with tf.Session() as sess3:
    print(sess3.run(b))
```

```python
#使用切片和索引取出图片的不同
#切片是原始数据的视图，而索引是取出的真实数据
```

```python
#取出1357张图片
for i in range (1,4):
    j=1
    c = tf.slice(img,[j-1,0,0,0],[1,28,28,3]) 
    j=j+2
    with tf.Session() as sess4:
        print(sess4.run(c))
```

```python
#第七张中心区域
d = tf.slice(img,[6,7,7,0],[1,14,14,3])
with tf.Session() as sess5:
    print(sess5.run(d))
```

```python
#三份单通道图片
e = tf.slice(img,[0,0,0,0],[10,28,28,1])
f = tf.slice(img,[0,0,0,1],[10,28,28,1])
g = tf.slice(img,[0,0,0,2],[10,28,28,1])
with tf.Session() as sess6:
    print(sess6.run(e))
with tf.Session() as sess7:
    print(sess7.run(f))
with tf.Session() as sess8:
    print(sess8.run(g))
```
